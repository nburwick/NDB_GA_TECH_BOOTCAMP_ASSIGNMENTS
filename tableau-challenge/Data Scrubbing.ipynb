```python
import pandas as pd
import numpy as np
import os
from fnmatch import fnmatch
import datetime as dt
from geopy import distance
import multiprocessing
from threading import Thread
```

```python
df = pd.read_csv(os.path.join('Resources','202304-citibike-tripdata.csv'),delimiter=',', low_memory=False)

df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

df
```

```python
print(f"{len(df):,}")
```

```python
start_stations_unique = df[['start_station_name', 'start_station_id',
                            'start_lat', 'start_lng']].drop_duplicates(subset=['start_station_name'])\
                            .dropna().sort_values('start_station_id').reset_index(drop=True).rename(columns={'start_station_name':'station_name',
                                                                                                             'start_station_id':'station_id', 
                                                                                                             'start_lat':'lat', 'start_lng':'lng'})
start_stations_unique
```

```python
end_stations_unique = df[['end_station_name', 'end_station_id',
                          'end_lat','end_lng']].dropna().drop_duplicates(subset=['end_station_name']) \
                            .sort_values('end_station_id').reset_index(drop=True).rename(columns={'end_station_name':'station_name',
                                                                                                  'end_station_id':'station_id', 
                                                                                                  'end_lat':'lat', 'end_lng':'lng'})
end_stations_unique
```

```python
stations_df = pd.merge(left=end_stations_unique, right=start_stations_unique, how='outer', indicator=True)
stations_df = stations_df.drop_duplicates(subset=['station_name'])
stations_df = stations_df.sort_values('station_id').reset_index(drop=True)
stations_df.dropna(inplace=True)

display(stations_df)

display(stations_df['_merge'].value_counts())
stations_df.to_csv(os.path.join('Cleaned Data', 'citi_bike_stations.csv'), index=False)
```

```python
df['trip_duration'] = (df.ended_at - df.started_at) / dt.timedelta(minutes=1)
```

```python
for col in df.columns:
    dtype = df[col].dtype.name
    if dtype == 'object':
        df[col] = df[col].astype('string')
```

```python
# Build Thread Pool
num_processes = 150

# Create Chunk Size
chunk_size = int(df.shape[0]/num_processes)

# Create Chunks
chunks = [df.iloc[df.index[i:i + chunk_size]] for i in range(0, df.shape[0], chunk_size)]

# Print number of chunks
print(f'There are {len(chunks)} chunks')
```

```python
chunks[0].shape[0]
```

```python
df_list = []
def calc_dist(chunk):
    chunk = chunk.reset_index(drop=True)
    chunk['est_trip_distance_miles'] = ''
    for index, row in chunk.iterrows():
        try:
            coords_1 = (row['start_lat'], row['start_lng'])
            coords_2 = (row['end_lat'],row['end_lng'])
            chunk.loc[index, 'est_trip_distance_miles'] = round(distance.geodesic(coords_1, coords_2).miles,2)
        except:
            chunk.loc[index, 'est_trip_distance_miles'] = np.nan
    df_list.append(chunk)
```

```python
%%time

# Empty lists
threads = []

for chunk in chunks:
    thread = Thread(target=calc_dist, kwargs={'chunk' : chunk})
    thread.start()
    threads.append(thread)
    
for thread in threads:
    thread.join()
```

```python
df = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
df.dropna(inplace=True)
df['est_trip_distance_miles'] = df['est_trip_distance_miles'].astype('float64')
```

```python
start_stations = pd.DataFrame({'values' : df['start_station_name'].value_counts()}).reset_index(names='start_station_name')
start_stations
```

```python
end_stations = pd.DataFrame({'count_of_end_destinations' : df['end_station_name'].value_counts()})
end_stations.index.names = ['station_name']
end_stations
```

```python
avg_trips_by_start_station = df[['start_station_name', 'end_station_name', 'trip_duration', 'est_trip_distance_miles']].groupby('start_station_name')\
                                .agg({'start_station_name':'count', 'trip_duration':'mean', 'est_trip_distance_miles': 'mean'})\
                                .rename(columns={'start_station_name': 'count_of_starts'})
avg_trips_by_start_station.index.names = ['station_name']
avg_trips_by_start_station
```

```python
stations_analysis = pd.merge(left=avg_trips_by_start_station, right=end_stations, how='outer', right_index=True, left_index=True)
stations_analysis
```

```python
stations_analysis.to_csv(os.path.join('Cleaned Data', 'individual_station_averages.csv'))
```

```python
%%time
df['route'] = df[['start_station_name', 'end_station_name']].apply(lambda x : " - ".join(sorted(x)), axis=1)
df['route_one_way'] = df[['start_station_name', 'end_station_name']].apply(" - ".join, axis=1)
df
```

```python
one_way_df = df[['route_one_way', 'trip_duration', 'est_trip_distance_miles']].groupby('route_one_way')\
                                .agg({'route_one_way':'count', 'trip_duration':'mean', 'est_trip_distance_miles': 'mean'})\
                                .rename(columns={'route_one_way': 'count_one_way_routes'})
one_way_df.index.names = ['one_way_route']
one_way_df.to_csv(os.path.join('Cleaned Data', 'one_way_routes.csv'))
one_way_df
```

```python
combined_routes_df = df[['route', 'trip_duration', 'est_trip_distance_miles']].groupby('route')\
                                .agg({'route':'count', 'trip_duration':'mean', 'est_trip_distance_miles': 'mean'})\
                                .rename(columns={'route': 'count_one_way_routes'})
combined_routes_df.index.names = ['combined_route']
combined_routes_df.to_csv(os.path.join('Cleaned Data', 'combined_routes.csv'))
combined_routes_df
```
