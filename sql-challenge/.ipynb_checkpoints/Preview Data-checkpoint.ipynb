```python
import os, pandas as pd, fnmatch
```

```python
folder = os.path.join('data')
```

```python
print(folder)
```

```python
csv_list = fnmatch.filter(os.listdir(folder), '*.csv')
```

```python
csv_list
```

```python
df_keys = [file.split('.')[0] for file in csv_list]
df_keys
```

```python
df_list = [pd.read_csv(os.path.join(folder,file), delimiter=',') for file in csv_list]
```

```python
df_list
```

```python
df_dict = dict(zip(df_keys,df_list))
```

```python
df_dict
```

```python
for key, item in df_dict.items():
    print(key)
    display(item.head())
```
